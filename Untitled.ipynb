* CAT es la acciÃ³n mas volÃ¡til, ya que tiene una desviaciÃ³n estÃ¡ndar de usd 6,22

### Ya que conocemos la acciÃ³n mÃ¡s vÃ³latil, podemos preguntarnos si el precio de esta acciÃ³n correlaciÃ³n con el precio de otras acciones

### Deseamos estudiar una o varias acciones particulares pero el conjunto de datos no presenta un formato conveniente para ello.

### La funciÃ³n df.pivot_table() nos genera un nuevo DataFrame reorganizado con la forma que deseamos

```python
df_nuevo = df.pivot_table(index='fecha',
                         columns= 'accion',
                         values= 'precio')
df_nuevo
```

## Ahora podemos centrarnos en estudiar el comportamiento de cada acciÃ³n en el tiempo

```python
fig, (pan1,pan2) = plt.subplots(2, 1, sharex=True, constrained_layout=True)
  
pan1.plot(df_nuevo.index, df_nuevo.CAT,
         color='blue',
         marker='o', 
         linestyle= 'solid',
         label='CAT')

pan2.plot(df_nuevo.index, df_nuevo.AA,
         color='orange',
         marker='v', 
         linestyle= 'solid',
         label='AA')

for pan in fig.get_axes():
    pan.set(xlabel='Fecha',
            ylabel='DÃ³lares americanos.',
            title='Precio de la acciÃ³n (al cierre).')
    
pan1.legend()
pan2.legend()
```

Los precios de ambas acciones presentan un comportamiento alcista hasta Abril - Con un ligero declive en las primeras semanas de marzo - A partir de ahÃ­ el precio baja hasta la Ãºltima semana de junio.

### Â¿Existe alguna correlaciÃ³n entre los precios de CAT y el resto de las acciones?

```python
df_nuevo.corr() #obtiene la matriz de correlacion entre las acciones. Y nos indica cuanto se mueve una acciÃ³n en funciÃ³n de otra acciÃ³n
```

```python
df_nuevo.corr().loc['CAT'] # metodo loc nos permite indexar. Nos mostrarÃ¡ las correlaciones de los valores de las acciones con la acciÃ³n CAT
```

```python
corr_pos = df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT']>0.8].sort_values(ascending=False)
corr_pos
```

Las acciones DD, VZ, CVX, MMM, XOM presentan una fuerte correlaciÃ³n positiva con CAT.
Cuando el coeficiente de correlacion es cercano a 1 podemos decir que la correlaciÃ³n entre ambas varianbles, es positiva, casi perfecta, es decir, aumentan en la misma proporciÃ³n

* Vamos a graficar los primeros cuatro elementos

```python
valores = corr_pos[1:5] # empieza en la posicion 1, saltea a CAT

plt.bar(valores.index, valores.values, color='forestgreen')

plt.plot()

plt.xlabel('Acciones')

plt.ylabel('Coeficiente de correlaciÃ³n.')

plt.title('CorrelaciÃ³n entre CAT')

plt.show()
```

#### CorrelaciÃ³n negativa

```python
corr_neg = df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT']<0].sort_values(ascending=True)
corr_neg
```

```python
valores = corr_neg[0:5]

plt.bar(valores.index, valores.values, color='brown')

plt.plot()

plt.xlabel('Acciones')

plt.ylabel('Coeficiente de correlaciÃ³n.')

plt.title('CorrelaciÃ³n entre CAT y el resto')

plt.show()
```

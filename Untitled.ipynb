## Cuáles acciones tuvieron la mayor y la menor variación de precios?

```python
df.groupby('accion').agg({'precio':[np.mean, np.std]}).sort_values([('precio', 'std')], ascending=False)
```

* CAT es la acción mas volátil, ya que tiene una desviación estándar de usd 6,22
* AA es la acción más consistente, con una desvicación estándar de usd 0,77

### Ya que conocemos la acción más vólatil, podemos preguntarnos si el precio de esta acción correlación con el precio de otras acciones

### Deseamos estudiar una o varias acciones particulares pero el conjunto de datos no presenta un formato conveniente para ello.

### La función df.pivot_table() nos genera un nuevo DataFrame reorganizado con la forma que deseamos

```python
df_nuevo = df.pivot_table(index='fecha',
                         columns= 'accion',
                         values= 'precio')
df_nuevo
```

## Ahora podemos centrarnos en estudiar el comportamiento de cada acción en el tiempo

```python
fig, (pan1,pan2) = plt.subplots(2, 1, sharex=True, constrained_layout=True)
  
pan1.plot(df_nuevo.index, df_nuevo.CAT,
         color='blue',
         marker='o', 
         linestyle= 'solid',
         label='CAT')

pan2.plot(df_nuevo.index, df_nuevo.AA,
         color='orange',
         marker='v', 
         linestyle= 'solid',
         label='AA')

for pan in fig.get_axes():
    pan.set(xlabel='Fecha',
            ylabel='Dólares americanos.',
            title='Precio de la acción (al cierre).')
    
pan1.legend()
pan2.legend()
```

Los precios de ambas acciones presentan un comportamiento alcista hasta Abril - Con un ligero declive en las primeras semanas de marzo - A partir de ahí el precio baja hasta la última semana de junio.

### ¿Existe alguna correlación entre los precios de CAT y el resto de las acciones?

```python
df_nuevo.corr() #obtiene la matriz de correlacion entre las acciones. Y nos indica cuanto se mueve una acción en función de otra acción
```

```python
df_nuevo.corr().loc['CAT'] # metodo loc nos permite indexar. Nos mostrará las correlaciones de los valores de las acciones con la acción CAT
```

```python
corr_pos = df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT']>0.8].sort_values(ascending=False)
corr_pos
```

Las acciones DD, VZ, CVX, MMM, XOM presentan una fuerte correlación positiva con CAT.
Cuando el coeficiente de correlacion es cercano a 1 podemos decir que la correlación entre ambas varianbles, es positiva, casi perfecta, es decir, aumentan en la misma proporción

* Vamos a graficar los primeros cuatro elementos

```python
valores = corr_pos[1:5] # empieza en la posicion 1, saltea a CAT

plt.bar(valores.index, valores.values, color='forestgreen')

plt.plot()

plt.xlabel('Acciones')

plt.ylabel('Coeficiente de correlación.')

plt.title('Correlación entre CAT')

plt.show()
```

#### Correlación negativa

```python
corr_neg = df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT']<0].sort_values(ascending=True)
corr_neg
```

```python
valores = corr_neg[0:5]

plt.bar(valores.index, valores.values, color='brown')

plt.plot()

plt.xlabel('Acciones')

plt.ylabel('Coeficiente de correlación.')

plt.title('Correlación entre CAT y el resto')

plt.show()
```

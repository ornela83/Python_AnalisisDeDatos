```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, pan_1 = plt.subplots(1, 1)

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)


pan_1.plot(x1, y1, linestyle = '--', color = 'g')
```

```python
fig, pan_1 = plt.subplots(1, 1)

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

pan_1.plot(x1, y1, linestyle = '--', color = 'y', marker = 'o', markeredgecolor = 'y', makerfacecolor = 'r')
```
